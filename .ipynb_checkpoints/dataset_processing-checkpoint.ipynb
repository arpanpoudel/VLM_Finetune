```python
import os
import re

def extract_findings_and_impression(report_text: str):
    text_upper = report_text.upper()
    findings_str = ""
    impression_str = ""

    findings_pattern = re.compile(r'FINDINGS:?\s*(.*?)(?=IMPRESSION:|$)', re.DOTALL)
    impression_pattern = re.compile(r'IMPRESSION:?\s*(.*?)(?=$)', re.DOTALL)

    f_match = findings_pattern.search(text_upper)
    if f_match:
        start_idx, end_idx = f_match.span(1)
        findings_str = report_text[start_idx:end_idx].strip()
    else:
        findings_str = "None"

    i_match = impression_pattern.search(text_upper)
    if i_match:
        start_idx, end_idx = i_match.span(1)
        impression_str = report_text[start_idx:end_idx].strip()
    else:
        impression_str = "None"

    return findings_str, impression_str

def process_reports(input_root: str, output_root: str):
    for dirpath, _, filenames in os.walk(input_root):
        for file in filenames:
            if file.endswith(".txt"):
                input_file_path = os.path.join(dirpath, file)

                # Compute relative path and corresponding output path
                relative_path = os.path.relpath(input_file_path, input_root)
                output_file_path = os.path.join(output_root, relative_path)

                # Create output directory if it doesn't exist
                os.makedirs(os.path.dirname(output_file_path), exist_ok=True)

                # Read input file
                with open(input_file_path, "r", encoding="utf-8") as f:
                    report_text = f.read()

                # Apply extraction
                findings, impression = extract_findings_and_impression(report_text)

                # Prepare content to save
                output_text = ""
                if findings:
                    output_text += "FINDINGS:\n" + findings + "\n\n"
                if impression:
                    output_text += "IMPRESSION:\n" + impression + "\n"

                # Save result to output folder
                with open(output_file_path, "w", encoding="utf-8") as out_f:
                    out_f.write(output_text)

                print(f"Processed: {input_file_path} → {output_file_path}")

# Example usage
if __name__ == "__main__":
    input_root = "/home/arpanp/Fed_Vision_Language/report/files"        # your input directory
    output_root = "/home/arpanp/Fed_Vision_Language/report/files_preprocessed"      # desired output directory
    process_reports(input_root, output_root)
```

```python
import os
import shutil

def copy_txt_files_with_structure(source_root: str, destination_root: str):
    for dirpath, _, filenames in os.walk(source_root):
        for file in filenames:
            if file.endswith(".txt"):  # adjust if you want to copy other types
                source_path = os.path.join(dirpath, file)

                # Preserve relative path
                relative_path = os.path.relpath(source_path, source_root)
                destination_path = os.path.join(destination_root, relative_path)

                # Create destination directories as needed
                os.makedirs(os.path.dirname(destination_path), exist_ok=True)

                # Copy the file
                shutil.copy2(source_path, destination_path)

                print(f"Copied: {source_path} → {destination_path}")

# Example usage
if __name__ == "__main__":
    source_root = "/home/arpanp/Fed_Vision_Language/report/files_preprocessed"    # original folder with subfolders and .txt files
    destination_root = "/home/arpanp/Fed_Vision_Language/files"  # destination root to mirror structure and copy files
    copy_txt_files_with_structure(source_root, destination_root)
```

```python
import os
import shutil

def copy_cleaned_reports(cleaned_root: str, files_root: str):
    """
    Copy only those .txt reports that you successfully cleaned (i.e., that remain
    in cleaned_root) into the matching subfolders under files_root.

    - cleaned_root:    e.g. "/home/arpanp/.../report/qwen_report_cleaned"
    - files_root:      e.g. "/home/arpanp/.../Fed_Vision_Language/files"
    """
    for dirpath, _, filenames in os.walk(cleaned_root):
        for fname in filenames:
            if not fname.lower().endswith(".txt"):
                continue

            # full path to the cleaned report
            cleaned_path = os.path.join(dirpath, fname)

            # compute its path relative to cleaned_root, then mirror under files_root
            rel_path = os.path.relpath(cleaned_path, cleaned_root)
            dest_path = os.path.join(files_root, rel_path)

            # ensure the destination directory exists
            os.makedirs(os.path.dirname(dest_path), exist_ok=True)

            # copy the cleaned .txt over
            shutil.copy2(cleaned_path, dest_path)
            print(f"Copied cleaned report:\n  {cleaned_path}\n→ {dest_path}")

if __name__ == "__main__":
    cleaned_root = "/home/arpanp/Fed_Vision_Language/report/qwen_report_cleaned"
    files_root   = "/home/arpanp/Fed_Vision_Language/files"
    copy_cleaned_reports(cleaned_root, files_root)
```

```python
class ChestDataset(Dataset):
    def __init__(self, index_path, transform=None, max_images=2):
        with open(index_path, "r") as f:
            self.samples = json.load(f)
        self.transform = transform
        self.max_images = max_images

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        sample = self.samples[idx]
        images = []
        for img_path in sample["image_paths"][:self.max_images]:
            img = Image.open(img_path).convert("RGB")
            if self.transform:
                img = self.transform(img)
            images.append(img)
        with open(sample["text_path"], "r", encoding="utf-8") as f:
            text = f.read().strip()
        return images, text
           
```

```python
import os
import json

def build_index(root_dir, output_json="mimic_index.json"):
    samples = []
    for dirpath, _, filenames in os.walk(root_dir):
        for fname in filenames:
            if fname.endswith(".txt"):
                study_id = os.path.splitext(fname)[0]
                txt_path = os.path.join(dirpath, fname)
                img_dir = os.path.join(dirpath, study_id)
                if os.path.isdir(img_dir):
                    img_files = [os.path.join(img_dir, f)
                                 for f in sorted(os.listdir(img_dir))
                                 if f.lower().endswith(".jpg")]
                    if img_files:
                        samples.append({
                            "text_path": txt_path,
                            "image_paths": img_files
                        })
    with open(output_json, "w") as f:
        json.dump(samples, f)
    print(f"Indexed {len(samples)} samples.")

# Run once
build_index("files", "mimic_index_clean.json")
```

## Training set frontal index file

```python
import os
import json
import pandas as pd

# Load frontal metadata
meta_df = pd.read_csv("dataset_files/mimic-cxr-2.0.0-metadata.csv.gz")
frontal_df = meta_df[meta_df['ViewPosition'].isin(['PA', 'AP'])].copy()

# Build full relative image path
def build_path(row):
    return os.path.join(
        "files",
        f"p{str(row.subject_id).zfill(8)[:2]}",
        f"p{str(row.subject_id).zfill(8)}",
        f"s{row.study_id}",
        f"{row.dicom_id}.jpg"
    )

frontal_df['jpg_path'] = frontal_df.apply(build_path, axis=1)

# Load split file and filter for training set
split_df = pd.read_csv("dataset_files/mimic-cxr-2.0.0-split.csv.gz")
train_studies = set(split_df[split_df['split'] == 'validate']['study_id'])

# Keep only frontal images that are in the training set
frontal_df = frontal_df[frontal_df['study_id'].isin(train_studies)]

# Create a set of valid frontal image paths
frontal_paths_set = set(frontal_df['jpg_path'])

# Load original index
with open("index_files/mimic_index_clean.json", "r") as f:
    samples = json.load(f)

# Filter samples: frontal view images from training set only
filtered_samples = []
for sample in samples:
    study_id = int(os.path.basename(sample["text_path"]).replace(".txt", "").replace("s", ""))
    if study_id not in train_studies:
        continue
    filtered_image_paths = [p for p in sample["image_paths"] if p in frontal_paths_set]
    if filtered_image_paths:
        filtered_samples.append({
            "text_path": sample["text_path"],
            "image_paths": filtered_image_paths
        })

# Save the new filtered JSON
with open("index_files/mimic_index_frontal_valid.json", "w") as f:
    json.dump(filtered_samples, f, indent=2)

print(f"Saved {len(filtered_samples)} training samples with frontal views.")
```

```python
import os

def count_txt_files(root_dir: str) -> int:
    """
    Recursively count all .txt files under root_dir.
    """
    count = 0
    for dirpath, dirnames, filenames in os.walk(root_dir):
        for fname in filenames:
            if fname.lower().endswith(".txt"):
                count += 1
    return count

if __name__ == "__main__":
    root_folder = "/home/arpanp/Fed_Vision_Language/report/qwen_report_cleaned"
    total_txt = count_txt_files(root_folder)
    print(f"Total .txt files under '{root_folder}': {total_txt}")
```

```python
import pandas as pd
import os
import random
import shutil
from tqdm import tqdm

# Paths
csv_path = "/home/arpanp/Fed_Vision_Language/dataset_files/mimic-cxr-2.1.0-test-set-labeled.csv"
source_root = "/home/arpanp/Fed_Vision_Language/files"
target_root = "/home/arpanp/Fed_Vision_Language/test"

# Step 1: Load CSV and select random 100 study_ids
df = pd.read_csv(csv_path)
study_ids = df['study_id'].unique()
random_study_ids = random.sample(list(study_ids), 100)

# Step 2: Copy corresponding .txt and .jpg files
for study_id in tqdm(random_study_ids, desc="Copying files", ncols=80):
    study_file_prefix = f"s{int(study_id)}"
    found = False

    # Search for the study file in the directory tree
    for dirpath, _, filenames in os.walk(source_root):
        for fname in filenames:
            if fname.startswith(study_file_prefix):
                source_file_path = os.path.join(dirpath, fname)

                # Compute relative path and destination
                rel_path = os.path.relpath(source_file_path, source_root)
                target_file_path = os.path.join(target_root, rel_path)

                # Ensure destination directory exists
                os.makedirs(os.path.dirname(target_file_path), exist_ok=True)

                # Copy file
                shutil.copy2(source_file_path, target_file_path)
                found = True

    if not found:
        print(f"Warning: Files for study_id {study_id} not found!")

print("Copy completed.")
```
